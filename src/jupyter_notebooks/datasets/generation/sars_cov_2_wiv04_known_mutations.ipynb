```python
import pandas as pd
import os
import numpy as np
import re
```

```python
file_path = os.path.join(os.getcwd(), "..", "..", "..","..", "input/data/coronaviridae/20240313/wiv04/SARS_CoV_2_Variants_ViralZone.csv")
```

```python
df = pd.read_csv(file_path)
```

```python
df
```

```python
df["spike_mutations"] = df["spike_mutations"].apply(lambda x: x.split(", "))
```

```python
df
```

```python
df_mutations_exploded = df.explode("spike_mutations")
df_mutations_exploded
```

```python
mutations = df_mutations_exploded["spike_mutations"].unique()
mutations
```

```python
len(mutations)
```

```python
df_mutations_exploded["mutation_position"] = df_mutations_exploded["spike_mutations"].str.extract(r".(\d+).")
```

```python
df_mutations_exploded
```

```python
df_mutations_exploded.to_csv(os.path.join(os.getcwd(), "..", "..", "..","..", "input/data/coronaviridae/20240313/wiv04/SARS_CoV_2_WIV04_mutations.csv"), index=0)
```

```python
df_mutations_grp = df_mutations_exploded[["pango_lineage", "spike_mutations"]].groupby(["spike_mutations"]).agg({"pango_lineage": lambda x: x.tolist()}).reset_index()
df_mutations_grp
```

```python
df_mutations_grp["variant_count"] = df_mutations_grp["pango_lineage"].apply(lambda x: len(x))
df_mutations_grp
```

```python
df_mutations_grp.to_csv(os.path.join(os.getcwd(), "..", "..", "..","..", "input/data/coronaviridae/20240313/wiv04/SARS_CoV_2_WIV04_mutations_grouped.csv"), index=0)
```

```python
df_mutations_grp_filtered = df_mutations_grp[df_mutations_grp["spike_mutations"].str.match(r"[A-Z]\d+[A-Z]")]
df_mutations_grp_filtered
```

```python
df_mutations_grp_filtered.to_csv(os.path.join(os.getcwd(), "..", "..", "..","..", "input/data/coronaviridae/20240313/wiv04/SARS_CoV_2_WIV04_mutations_grouped_filtered.csv"), index=0)
```

```python
df_mutations_grp_filtered = df_mutations_grp_filtered.explode("pango_lineage")
df_mutations_grp_filtered["variant_count"] = df_mutations_grp_filtered["variant_count"].apply(lambda x: 1 if x >= 1 else 0)
df_mutations_grp_filtered.pivot(index="spike_mutations", columns="pango_lineage", values="variant_count").to_csv(os.path.join(os.getcwd(), "..", "..", "..","..", "input/data/coronaviridae/20240313/wiv04/SARS_CoV_2_WIV04_mutations_filtered_heatmap.csv"))
```

```python
df_mutation_positions_grp = df_mutations_exploded[["pango_lineage", "mutation_position"]].groupby(["mutation_position"]).agg({"pango_lineage": lambda x: x.tolist()}).reset_index()
df_mutation_positions_grp
```

```python
df_mutation_positions_grp["variant_count"] = df_mutation_positions_grp["pango_lineage"].apply(lambda x: len(x))
df_mutation_positions_grp
```

```python
df_mutation_positions_grp.to_csv(os.path.join(os.getcwd(), "..", "..", "..","..", "input/data/coronaviridae/20240313/wiv04/SARS_CoV_2_WIV04_mutation_positions_grouped.csv"), index=0)
```
