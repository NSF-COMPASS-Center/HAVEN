```python
import pandas as pd
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
sars_cov_2_variants_file_path = os.path.join(os.getcwd(), "..", "..", "..", "..", "input/data/coronaviridae/20240313/sarscov2/uniprot/variants/sarscov2_variants_s.csv")
cov_s_uniref90_file_path = os.path.join(os.getcwd(), "..", "..", "..", "..", "input/data/coronaviridae/20240313/uniref/coronaviridae_s_uniref90_embl_hosts_pruned_metadata_corrected_species_virus_host_vertebrates_w_seq_t0.01_c8.csv")
```

```python
sars_cov_2_variants_df = pd.read_csv(sars_cov_2_variants_file_path)
sars_cov_2_variants_df
```

```python
mapping_df = pd.read_csv(os.path.join(os.getcwd(), "..","..", "..", "..", "input/data/coronaviridae/20240313/wiv04/SARS_CoV_2_Variants_ViralZone.csv"))
sars_cov_2_variants_df = sars_cov_2_variants_df.merge(mapping_df[["seq", "pango_lineage", "emergence_date"]], how="left", left_on="id", right_on="seq").rename(columns={"seq_x": "seq"}).drop(columns=["seq_y"])
```

```python
sars_cov_2_variants_df
```

```python
cov_s_uniref90_df = pd.read_csv(cov_s_uniref90_file_path)
cov_s_uniref90_df
```

```python
from Bio import Align
aligner = Align.PairwiseAligner(match_score=1.0)
```

```python
wiv04_seq = sars_cov_2_variants_df[sars_cov_2_variants_df["id"] == "WIV04"]["seq"].values[0]
wiv04_seq
```

```python
sars_cov_2_variants_df["seq_sim_w_wiv04"] = sars_cov_2_variants_df["seq"].apply(lambda x: aligner.score(wiv04_seq, x) / len(wiv04_seq) * 100)
```

```python
sars_cov_2_variants_df[["pango_lineage", "seq_sim_w_wiv04"]].rename(columns={"pango_lineage": "Variant", "seq_sim_w_wiv04": "Similarity with WIV04 (%)"})
```
